# Вычисление метрик (для себя)

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Линейная модель
logistic = LogisticRegression(max_iter=1000)
logistic.fit(X, y)
y_pred_lin = logistic.predict(X)

print("=== Линейная модель ===")
print("Accuracy:", round(accuracy_score(y, y_pred_lin), 4))
print("\nClassification Report:\n", classification_report(y, y_pred_lin))
# Матрица ошибок
sns.heatmap(confusion_matrix(y, y_pred_lin), annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Linear")
plt.show()
```

```python
# Полиномиальная модель (degree=2) 

from sklearn.preprocessing import PolynomialFeatures

# Создаём полиномиальные признаки 2-й степени
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Обучаем логистическую регрессию на полиномиальных признаках
polynomial = LogisticRegression(max_iter=1000)
polynomial.fit(X_poly, y)
y_pred_poly = polynomial.predict(X_poly)

print("=== Полиномиальная модель (degree=2) ===")
print("Accuracy:", round(accuracy_score(y, y_pred_poly), 4))
print("\nClassification Report:\n", classification_report(y, y_pred_poly))

# Матрица ошибок
sns.heatmap(confusion_matrix(y, y_pred_poly), annot=True, fmt='d', cmap='Greens')
plt.title("Confusion Matrix - Polynomial (deg=2)")
plt.show()
```

```python
import pandas as pd

comparison = pd.DataFrame({
    'Модель': ['Линейная', 'Полиномиальная (deg=2)'],
    'Accuracy': [accuracy_score(y, y_pred_lin), accuracy_score(y, y_pred_poly)]
})
print(comparison.round(4))
```

# Задачи для самостоятельного решения

## Задание 1: Дополнительная информация о LogisticRegression, влияние аргументов

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

base = LogisticRegression(max_iter=1000).fit(X, y)

# Дополнительная информация о модели
print("classes_:    ", base.classes_)
print("n_iter_:     ", base.n_iter_)          # сколько итераций потребовалось
print("n_features_: ", base.n_features_in_)
print("intercept_:  ", base.intercept_.round(4))

# Влияние аргументов: C (регуляризация), solver, penalty
for C, solver, penalty in [
    (0.01, 'lbfgs',  'l2'),   # сильная регуляризация
    (1.0,  'lbfgs',  'l2'),   # по умолчанию
    (100,  'lbfgs',  'l2'),   # слабая регуляризация
    (1.0,  'saga',   'l1'),   # L1 (разреженные веса)
]:
    m = LogisticRegression(C=C, solver=solver, penalty=penalty, max_iter=1000).fit(X, y)
    print(f"C={C:>5}, solver={solver}, penalty={penalty}  -> accuracy={m.score(X,y):.4f}")
```

## Задание 2–3: Все модели классификации + функция авто-обучения

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import PassiveAggressiveClassifier, RidgeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, f1_score
import seaborn as sns

MODELS = {
    "LogisticRegression":       LogisticRegression(max_iter=1000),
    "SVC (linear)":             make_pipeline(StandardScaler(), SVC(kernel='linear')),
    "SVC (rbf)":                make_pipeline(StandardScaler(), SVC(kernel='rbf')),
    "SVC (poly)":               make_pipeline(StandardScaler(), SVC(kernel='poly')),
    "SVC (no kernel / linear)": make_pipeline(StandardScaler(), SVC(kernel='linear', C=1)),
    "KNeighbors":               KNeighborsClassifier(),
    "MLP":                      make_pipeline(StandardScaler(), MLPClassifier(max_iter=500, random_state=0)),
    "DecisionTree":             DecisionTreeClassifier(random_state=0),
    "GaussianNB":               GaussianNB(),
    "PassiveAggressive":        make_pipeline(StandardScaler(), PassiveAggressiveClassifier(max_iter=1000, random_state=0)),
    "RidgeClassifier":          make_pipeline(StandardScaler(), RidgeClassifier()),
    "RandomForest":             RandomForestClassifier(n_estimators=100, random_state=0),
    "Bagging":                  BaggingClassifier(n_estimators=50, random_state=0),
}


def evaluate_classifiers(X, y, models=MODELS):
    """Задание 3: обучает все модели, возвращает DataFrame с метриками."""
    rows = []
    for name, m in models.items():
        m.fit(X, y)
        yp = m.predict(X)
        rows.append({
            "модель":    name,
            "accuracy":  accuracy_score(y, yp),
            "f1":        f1_score(y, yp, average='weighted'),
        })
    return pd.DataFrame(rows).set_index("модель").round(4)


results_diab = evaluate_classifiers(X, y)
results_diab.sort_values("accuracy", ascending=False)
```

```python
# Матрицы ошибок для всех моделей
fig, axes = plt.subplots(4, 4, figsize=(16, 14))
axes = axes.ravel()

for ax, (name, m) in zip(axes, MODELS.items()):
    yp = m.predict(X)
    cm = confusion_matrix(y, yp)
    sns.heatmap(cm, annot=True, fmt='d', ax=ax, cbar=False)
    ax.set_title(f"{name}\nacc={accuracy_score(y,yp):.3f}", fontsize=7)
    ax.set_xlabel("predicted", fontsize=7)
    ax.set_ylabel("true", fontsize=7)

for ax in axes[len(MODELS):]:
    fig.delaxes(ax)

plt.tight_layout()
plt.show()
```

## Задание 4: Повтор анализа — датасет Iris

```python
from sklearn.datasets import load_iris

iris = load_iris()
X_ir = pd.DataFrame(iris.data, columns=iris.feature_names)
y_ir = pd.Series(iris.target, name='species')

print(X_ir.shape)
X_ir.describe().round(2)
```

```python
# Базовая модель + отчёт
base_ir = LogisticRegression(max_iter=1000).fit(X_ir, y_ir)
yp_ir = base_ir.predict(X_ir)
print(classification_report(y_ir, yp_ir, target_names=iris.target_names))

sns.heatmap(confusion_matrix(y_ir, yp_ir), annot=True, fmt='d',
            xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Iris — LogisticRegression")
plt.show()
```

```python
# Полиномиальная модель deg=2
from sklearn.preprocessing import PolynomialFeatures
X_ir_poly = PolynomialFeatures(2).fit_transform(X_ir)
poly_ir = LogisticRegression(max_iter=1000).fit(X_ir_poly, y_ir)
print("Poly deg=2 accuracy:", round(poly_ir.score(X_ir_poly, y_ir), 4))
```

```python
# Все модели на Iris (пересоздаём экземпляры)
MODELS_IR = {
    "LogisticRegression":  LogisticRegression(max_iter=1000),
    "SVC (linear)":        make_pipeline(StandardScaler(), SVC(kernel='linear')),
    "SVC (rbf)":           make_pipeline(StandardScaler(), SVC(kernel='rbf')),
    "SVC (poly)":          make_pipeline(StandardScaler(), SVC(kernel='poly')),
    "KNeighbors":          KNeighborsClassifier(),
    "MLP":                 make_pipeline(StandardScaler(), MLPClassifier(max_iter=500, random_state=0)),
    "DecisionTree":        DecisionTreeClassifier(random_state=0),
    "GaussianNB":          GaussianNB(),
    "PassiveAggressive":   make_pipeline(StandardScaler(), PassiveAggressiveClassifier(max_iter=1000, random_state=0)),
    "RidgeClassifier":     make_pipeline(StandardScaler(), RidgeClassifier()),
    "RandomForest":        RandomForestClassifier(n_estimators=100, random_state=0),
    "Bagging":             BaggingClassifier(n_estimators=50, random_state=0),
}

results_ir = evaluate_classifiers(X_ir, y_ir, models=MODELS_IR)
results_ir.sort_values("accuracy", ascending=False)
```

```python
# Матрицы ошибок для Iris
fig, axes = plt.subplots(3, 4, figsize=(16, 11))
axes = axes.ravel()

for ax, (name, m) in zip(axes, MODELS_IR.items()):
    yp = m.predict(X_ir)
    sns.heatmap(confusion_matrix(y_ir, yp), annot=True, fmt='d', ax=ax, cbar=False)
    ax.set_title(f"{name}\nacc={accuracy_score(y_ir,yp):.3f}", fontsize=7)

for ax in axes[len(MODELS_IR):]:
    fig.delaxes(ax)

plt.suptitle("Iris dataset", fontsize=12)
plt.tight_layout()
plt.show()

# Итоговое сравнение двух датасетов
print("=== Diabetes ===")
print(results_diab.sort_values("accuracy", ascending=False).to_string())
print("\n=== Iris ===")
print(results_ir.sort_values("accuracy", ascending=False).to_string())
```
