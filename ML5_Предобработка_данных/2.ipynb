# Задачи для самостоятельрого решения


## Задание 1. Другие методы визуализации распределений

```python
num_cols = ['SO2', 'NO2', 'CO', 'O3', 'PRES', 'WSPM']

# Boxplot — наглядно показывает выбросы и квартили
fig, axes = plt.subplots(2, 3, figsize=(14, 7))
for ax, col in zip(axes.flat, num_cols):
    ax.boxplot(prsa_data[col].dropna(), vert=True, patch_artist=True)
    ax.set_title(col); ax.set_xlabel(col)
plt.suptitle('Boxplot числовых признаков')
plt.tight_layout(); plt.show()
```

```python
# Violin plot — boxplot + KDE, виден полный профиль распределения
fig, axes = plt.subplots(2, 3, figsize=(14, 7))
for ax, col in zip(axes.flat, num_cols):
    ax.violinplot(prsa_data[col].dropna())
    ax.set_title(col)
plt.suptitle('Violin plot числовых признаков')
plt.tight_layout(); plt.show()
```

```python
# Совместное распределение с AQI Label — catplot (boxen) вместо kdeplot
# Показывает медиану, квартили и форму распределения по классам
fig, axes = plt.subplots(2, 3, figsize=(15, 9))
for ax, col in zip(axes.flat, num_cols):
    data_plot = prsa_data[[col, 'AQI Label']].dropna()
    groups = [data_plot[col][data_plot['AQI Label'] == lbl].values
              for lbl in data_plot['AQI Label'].unique()]
    labels = data_plot['AQI Label'].unique()
    ax.boxplot(groups, labels=labels, patch_artist=True)
    ax.set_title(col); ax.tick_params(axis='x', rotation=30)
plt.suptitle('Признаки по классам AQI (boxplot)')
plt.tight_layout(); plt.show()
# Вывод: SO2, NO2, CO явно разделяют классы → информативны для модели
```

## Задание 2. Адаптивный клиппинг через процентили

```python
# Функция клиппинга по нижнему и верхнему процентилям
def adaptive_clip(df, col, low_pct=1, high_pct=99):
    lo = df[col].quantile(low_pct / 100)
    hi = df[col].quantile(high_pct / 100)
    df[col] = df[col].clip(lo, hi)
    print(f"{col}: clip [{lo:.2f}, {hi:.2f}]")
    return df

clip_cols = ['NO2', 'CO', 'O3', 'WSPM']
for col in clip_cols:
    prsa_data = adaptive_clip(prsa_data, col)

# Проверяем результат
prsa_data[clip_cols].describe().loc[['min', 'max']]
```

## Задание 3. Заполнение оставшихся пропусков

```python
print("Пропуски до заполнения:")
print(prsa_data.isnull().sum()[prsa_data.isnull().sum() > 0])

# NO2 и CO — правосторонние распределения → медиана лучше среднего
prsa_data['NO2'] = prsa_data['NO2'].fillna(prsa_data['NO2'].median())
prsa_data['CO']  = prsa_data['CO'].fillna(prsa_data['CO'].median())

# Линейная интерполяция — временной ряд, соседние значения близки
prsa_data['O3']  = prsa_data['O3'].interpolate(method='linear')

print("Пропуски после заполнения:")
print(prsa_data.isnull().sum()[prsa_data.isnull().sum() > 0])
```

## Задание 4. Нормализация числовых признаков

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

# SO2 — уже логарифмирован, близок к нормальному → StandardScaler
# NO2, CO — после клиппинга, лёгкая правосторонность → RobustScaler (устойчив к остаткам выбросов)
# PRES  — близко к нормальному, узкий диапазон → StandardScaler
# O3, WSPM — правосторонние → MinMaxScaler (уже ограничены клиппингом)

scale_map = {
    'SO2':  StandardScaler(),
    'NO2':  RobustScaler(),
    'CO':   RobustScaler(),
    'PRES': StandardScaler(),
    'O3':   MinMaxScaler(),
    'WSPM': MinMaxScaler(),
}

prsa_scaled = prsa_data.copy()
for col, scaler in scale_map.items():
    prsa_scaled[col] = scaler.fit_transform(prsa_data[[col]])

prsa_scaled[list(scale_map)].describe().round(3)
```

## Задание 5. Коррелограмма числовых признаков

```python
num_for_corr = ['SO2', 'NO2', 'CO', 'O3', 'PRES', 'WSPM']
corr = prsa_data[num_for_corr].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, linewidths=0.5)
plt.title('Коррелограмма числовых признаков')
plt.tight_layout(); plt.show()

# Значимые корреляции (|r| > 0.4):
strong = [(c1, c2, round(corr.loc[c1,c2],2))
          for i,c1 in enumerate(num_for_corr)
          for c2 in num_for_corr[i+1:]
          if abs(corr.loc[c1,c2]) > 0.4]
print("Сильные корреляции (|r|>0.4):")
for r in strong: print(f"  {r[0]} — {r[1]}: {r[2]}")
# Вывод: SO2/NO2/CO положительно коррелируют (загрязнители одного источника)
# O3 отрицательно коррелирует с ними (фотохимическая реакция)
```

## Задание 6. Визуализация связей между признаками

```python
# Pairplot — scatter matrix со всеми парами признаков
sample = prsa_scaled[num_for_corr + ['AQI Label']].dropna().sample(2000, random_state=42)
sns.pairplot(sample, hue='AQI Label', plot_kws={'alpha': 0.3, 's': 10},
             diag_kind='kde')
plt.suptitle('Pairplot признаков по классам AQI', y=1.01)
plt.show()
# Вывод:
# SO2-NO2, SO2-CO, NO2-CO — линейная зависимость (общий источник — сжигание топлива)
# O3 vs CO/NO2 — нелинейная отрицательная связь
# PRES и WSPM слабо связаны с остальными
```

```python
# Heatmap средних значений признаков по классам AQI
aqi_means = prsa_data.groupby('AQI Label')[num_for_corr].mean()
plt.figure(figsize=(9, 4))
sns.heatmap(aqi_means.T, annot=True, fmt='.1f', cmap='YlOrRd', linewidths=0.5)
plt.title('Средние значения признаков по классам AQI')
plt.tight_layout(); plt.show()
# Вывод: SO2, NO2, CO растут вместе с классом загрязнения → наиболее значимые признаки
```
